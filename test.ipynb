```python
import pprint
from parser import TelegramParser
```

```python
# ユーザーがアップロードしたファイル名
# このファイルがスクリプトと同じディレクトリにあることを想定
FILE_NAME = "refs/1_20251001030733217_0011702434_94096424-01_01_test.txt"

try:
    print(f"'{FILE_NAME}' の解析を開始します...")
    
    # 1. パーサーのインスタンスを作成
    parser = TelegramParser(FILE_NAME, encoding='cp932')
    
    # 2. 解析を実行
    telegram_data = parser.parse()
    
    print("\n--- 解析成功 ---")
    
    # 3. DotDict 形式でのアクセス例
    print("\n[DotDict アクセス例]")
    print(f"  患者氏名: {telegram_data.content.patient_info.kanja_kanji_shimei}")
    print(f"  患者住所: {telegram_data.content.patient_info.kanja_juusho}")
    print(f"  依頼医名: {telegram_data.content.order_info.irai_i_info.name}")
    
    # 項目情報 (リスト) へのアクセス
    item_count = telegram_data.content.item_info.item_count
    print(f"\n  項目数: {item_count}")
    
    if item_count > 0:
        # 最初の項目
        item_1 = telegram_data.content.item_info.item_group[0]
        print(f"  項目1 名前: {item_1.name}")
        print(f"  項目1 数量: {item_1.suuryou}")
        # 2番目の項目（ブドウ糖注）
        item_2 = telegram_data.content.item_info.item_group[1]
        print(f"  項目2 名前: {item_2.name}")
        print(f"  項目2 数量: {item_2.suuryou}")

    # 4. dict全体の確認 (pprintを使用)
    # print("\n--- 全データ (DotDict) ---")
    # pprint.pprint(telegram_data)

except FileNotFoundError:
    print(f"実行エラー: {FILE_NAME} が見つかりません。")
except ValueError as ve:
    print(f"検証エラー: {ve}")
except Exception as e:
    print(f"不明なエラーが発生しました: {e}")
```
